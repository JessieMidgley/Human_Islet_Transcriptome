```r
markers_df <- FindMarkers(vasa_seq_ND, ident.1 = "Alpha", ident.2 = "Beta")
```

```r
markers_df
```

```r
markers_df$gene <- rownames(markers_df)

# Add significance category
markers_df$significance <- "Not Significant"
markers_df$significance[markers_df$avg_log2FC > 0.25 & markers_df$p_val_adj < 0.05] <- "Up"
markers_df$significance[markers_df$avg_log2FC < -0.25 & markers_df$p_val_adj < 0.05] <- "Down"
```

```r
markers_df
```

```r
nrow(markers_df)
```

```r
library(ggplot2)
library(ggrepel)

options(repr.plot.width = 10, repr.plot.height = 10)

top_genes <- markers_df[order(markers_df$p_val_adj), ][1:10, ]


ggplot(markers_df, aes(x = avg_log2FC, y = -log10(p_val_adj), color = significance)) +
  geom_point(alpha = 0.7, size = 1.5) +
  geom_text_repel(data = top_genes, aes(label = gene), size = 4, max.overlaps = 20) +

  scale_color_manual(values = c("Up" = "red", "Down" = "blue", "Not Significant" = "grey")) +
  theme_minimal() +
  xlab("Log2 Fold Change") +
  ylab("-log10 Adjusted P-value") 
```

```r
any(markers_df$p_val_adj == 0)   
```
