```python
from matplotlib.patches import Patch


for transcript_id in MYO3A_tx:

    frame_colors = {
        "0": "green",
    }

    exon_colors = {
        "Non-overlapping": "purple",
        "Unannotated": "crimson",
        "Annotated": "steelblue"
    }
    
    
    exons = orfanage_output_with_CDS[(orfanage_output_with_CDS["feature"] == "exon") & (orfanage_output_with_CDS["transcript_id"] == transcript_id)]
    cds = orfanage_output_with_CDS[(orfanage_output_with_CDS["feature"] == "CDS") & (orfanage_output_with_CDS["transcript_id"] == transcript_id) ]
    
    
    # Get ref transcript
    ref_id = combined_df[combined_df["transcript_id"] == transcript_id]["ref_id"].iloc[0]
    ref_transcript = gencode_v47[gencode_v47['transcript_id'] == ref_id]
    ref_transcript = ref_transcript[["start", "end", "transcript_id"]]
    ref_transcript["start"] = ref_transcript["start"].astype(int)
    ref_transcript["end"] = ref_transcript["end"].astype(int)
    
    
    fig, ax = plt.subplots(figsize=(11, 3))
    
    
    
    # Plot ref
    for _, row in ref_transcript.iterrows():
        ax.add_patch(
            plt.Rectangle(
                (row["start"], 0.8), row["end"] - row["start"], 0.3,
                edgecolor="grey",
                facecolor="grey"
            )
        )
    
    
    # Plot exons
    for _, row in exons.iterrows():
        if row["is_non_overlapping"]:
            color = "purple"
        elif row["is_unannotated"]:
            color = "crimson"
        else:
            color = "steelblue"
    
        ax.add_patch(
            plt.Rectangle(
                (row["Start"], 0.4), row["End"] - row["Start"], 0.3,
                edgecolor=color,
                facecolor=color
            )
        )
    
    
    # Plot ORFanage CDS 
    for _, row in cds.iterrows():
        color = frame_colors.get(row["frame"])
        ax.add_patch(
            plt.Rectangle(
                (row["Start"], 0.0), row["End"] - row["Start"], 0.3,
                edgecolor=color,
                facecolor=color
            )
        )
    
    
    cds_legend_elements = [
        Patch(facecolor=color, edgecolor=color, label=f'Frame {frame}')
        for frame, color in frame_colors.items()
    ]
    
    
    
    exon_legend_elements = [
        Patch(facecolor=color, edgecolor=color, label=label)
        for label, color in exon_colors.items()
    ]
    exon_legend = ax.legend(handles=exon_legend_elements, loc='lower left', bbox_to_anchor=(1, 0.5), title='Exon Status')
    cds_legend = ax.legend(handles=cds_legend_elements, loc='upper left', bbox_to_anchor=(1, 0.5), title='ORFanage CDS')
    ax.add_artist(exon_legend) 
    
    ax.set_xlim(exons["Start"].min() - 500, exons["End"].max() + 500)
    # ax.set_xlim(21000000, 21200000)
    ax.set_ylim(-0.2, 1.2)
    ax.set_yticks([0.15, 0.55, 0.95])
    ax.set_yticklabels(["ORFanage CDS", f"HIT transcript: {transcript_id}", f"Reference transcript: {ref_id}"])
    ax.set_xlabel("Genomic Position")
    plt.title(f"{combined_df.loc[combined_df['transcript_id'] == transcript_id, 'Gene name'].iloc[0]} ({combined_df.loc[combined_df['transcript_id'] == transcript_id, 'Strand'].iloc[0]})")
    
    plt.tight_layout()
    plt.show()
```

## Do any of these genes have an unannotated TSS?

```python
combined_df.head()
```

```python
combined_df["TSS_rank"] = combined_df.groupby("Gene stable ID")["Relative TSS usage"].rank(method="dense", ascending=False)

def get_TSS_type(row):
    if row["Relative TSS usage"] == 1:
        return "unique TSS"
    elif row["TSS_rank"] == 1:
        return "main TSS"
    elif row["TSS_rank"] == 2:
        return "secondary TSS"
    else:
        return "minor TSS"

combined_df["Ranked_TSS_type"] = combined_df.apply(get_TSS_type, axis=1)
combined_df.head()
```

```python
combined_df[(combined_df["Gene name"].isin(t2d_gene_names)) & 
            (combined_df["TSS distance category"] == ">100bp")]["Gene name"].nunique()
```

```python
novel_promoter_genes = set(combined_df[(combined_df["Gene name"].isin(t2d_gene_names)) & 
            (combined_df["TSS distance category"] == ">100bp")]["Gene name"].unique())
```

```python
df_tss = combined_df[combined_df["Gene name"].isin(t2d_gene_names)]

# Define annotation status
df_tss["annotation_status"] = df_tss["TSS distance category"].apply(
    lambda x: "Unannotated TSS" if x == ">100bp" else "Annotated TSS"
)

# Count TSS events by category and annotation status
counts = (
    df_tss.groupby(["Ranked_TSS_type", "annotation_status"])["peak_name"]
    .nunique()
    .reset_index(name="count")
)

# Pivot for plotting
pivot = counts.pivot(index="Ranked_TSS_type", columns="annotation_status", values="count").fillna(0)

# --- Plot ---
pivot = pivot.loc[["main TSS", "secondary TSS", "minor TSS"]]  # consistent order

pivot.plot(kind="bar", figsize=(5,4))
plt.ylabel("Number of TSS peaks")
plt.xlabel("TSS category")
plt.title("Distribution of TSS categories in\nmonogenic diabetes-associated genes")
plt.xticks(rotation=0)
plt.legend(loc="upper center", frameon=False)

plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3

colors = ['#66c2a5', '#fc8d62', '#8da0cb']

plt.figure(figsize=(3,3))
v = venn3(
    [novel_first_exon_genes, novel_promoter_genes, new_coding_exon_genes],
    set_labels=('Novel first exons', 'Novel promoters', 'Novel coding exons'),
    set_colors=colors,  
    alpha=1
)

# Style set labels
for text in v.set_labels:
    text.set_fontsize(10)

# Auto-center subset labels
subset_ids = ['100', '010', '110', '001', '101', '011', '111']

for idx, subset_id in enumerate(subset_ids):
    text = v.get_label_by_id(subset_id)
    if text:
        text.set_fontsize(10)
        # Get the subset patch (circle or intersection)
        patch = v.get_patch_by_id(subset_id)
        if patch:
            # Use the patch center for the label
            text.set_position(patch.center)

# Remove outer box and tighten layout
plt.box(False)
plt.tight_layout()
plt.show()
```
