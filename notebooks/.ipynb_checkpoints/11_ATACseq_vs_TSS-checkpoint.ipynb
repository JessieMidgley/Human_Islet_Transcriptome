# Overlap differentially expressed TSSs with ATACseq peaks

```python
import pandas as pd
import pyranges as pr
import numpy as np
import pyreadr
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager
import seaborn as sns
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

font_path = "../data_raw/fonts/LatinmodernmathRegular.otf"  
font_manager.fontManager.addfont(font_path)

mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
font = {'family' : 'Latin Modern Math',
       'weight' : 'normal',
       'size'   : 10}

mpl.rc('font', **font)
mpl.rc('lines', lw=1, c='k')
sns.set_context("notebook", font_scale=1, rc={"axes.linewidth": 1,
                                              "lines.linewidth": 1, 
                                              "font.size": 10,
                                              "axes.titlesize": 10,
                                              "axes.labelsize": 10,
                                              "legend.fontsize": 10,
                                              "xtick.labelsize": 8,
                                              "ytick.labelsize": 8})
```

```python
# Load ATACseq peaks (alpha vs beta)
atac_peaks = pd.read_csv("../data_raw/markersPeaks_ND_Beta_Alpha.bed", sep="\t", header=0)
atac_peaks.head()
```

```python
significant_peaks = atac_peaks[atac_peaks["FDR"] < 0.05]
```

```python
# Split significant peaks into alpha and beta
signifcant_alpha_peaks = significant_peaks[significant_peaks["MeanDiff"] < 0]
alpha_peak_pr = pr.PyRanges(
    signifcant_alpha_peaks.rename(columns={
        "seqnames": "Chromosome",
        "start": "Start",
        "end": "End"
    })[["Chromosome", "Start", "End", "FDR", "Log2FC", "idx"]]
)

signifcant_beta_peaks = significant_peaks[significant_peaks["MeanDiff"] > 0]
beta_peak_pr = pr.PyRanges(
    signifcant_beta_peaks.rename(columns={
        "seqnames": "Chromosome",
        "start": "Start",
        "end": "End"
    })[["Chromosome", "Start", "End", "FDR", "Log2FC", "idx"]]
)
```

```python
alpha_tss = pd.read_csv("../data_processed/VASAseq/alpha_tss.csv")
beta_tss = pd.read_csv("../data_processed/VASAseq/beta_tss.csv")
```

```python
# load full df to get TSS start and end
combined_df = pd.read_csv( '../data_processed/curated_HIT.tsv', sep='\t')
TSS_coords = combined_df[["TSS_name", "chr", "Strand", "TSS_start", "TSS_end"]].drop_duplicates("TSS_name")
TSS_coords["TSS_size"] = TSS_coords["TSS_end"] - TSS_coords["TSS_start"]
TSS_coords.head()
```

```python
alpha_tss = alpha_tss.merge(
    TSS_coords[["TSS_name", "chr", "Strand", "TSS_start", "TSS_end", "TSS_size"]],
    on="TSS_name",
    how="left"
)
alpha_tss.head()
```

```python
beta_tss = beta_tss.merge(
    TSS_coords[["TSS_name", "chr", "Strand", "TSS_start", "TSS_end", "TSS_size"]],
    on="TSS_name",
    how="left"
)
beta_tss.head()
```

### Alpha TSSs

```python
# Do these TSS overlap with significant alpha ATACseq peaks?
tss_pr = pr.PyRanges(
    alpha_tss.rename(columns={
        "chr": "Chromosome",
        "TSS_start": "Start",
        "TSS_end": "End"
    })[["Chromosome", "Start", "End", 
       "Gene.name", "TSS_name", "TSS_size"]]
)

# Find overlaps
overlaps = tss_pr.join(alpha_peak_pr)
alpha_TSS_overlap_df = overlaps.df
```

```python
# Compute overlap proportion
alpha_TSS_overlap_df["overlap_bp"] = (
    np.minimum(alpha_TSS_overlap_df["End"],   alpha_TSS_overlap_df["End_b"])
    - np.maximum(alpha_TSS_overlap_df["Start"], alpha_TSS_overlap_df["Start_b"])
).clip(lower=0)

alpha_TSS_overlap_df["overlap_proportion"] = alpha_TSS_overlap_df["overlap_bp"] / alpha_TSS_overlap_df["TSS_size"]
alpha_TSS_overlap_df
```

```python
# Distribution of overlap proportions
plt.figure(figsize=(6,4))
plt.hist(alpha_TSS_overlap_df["overlap_proportion"], bins=30, edgecolor="black")
plt.xlabel("Overlap proportion (Overlap length / TSS length)")
plt.ylabel("Count")
plt.show()
```

### Beta TSSs

```python
# Do these TSS overlap with significant beta ATACseq peaks?
tss_pr = pr.PyRanges(
    beta_tss.rename(columns={
        "chr": "Chromosome",
        "TSS_start": "Start",
        "TSS_end": "End"
    })[["Chromosome", "Start", "End", 
       "Gene.name", "TSS_name", "TSS_size"]]
)

# Find overlaps
overlaps = tss_pr.join(beta_peak_pr)
beta_TSS_overlap_df = overlaps.df
```

```python
# Compute overlap proportion
beta_TSS_overlap_df["overlap_bp"] = (
    np.minimum(beta_TSS_overlap_df["End"],   beta_TSS_overlap_df["End_b"])
    - np.maximum(beta_TSS_overlap_df["Start"], beta_TSS_overlap_df["Start_b"])
).clip(lower=0)

beta_TSS_overlap_df["overlap_proportion"] = beta_TSS_overlap_df["overlap_bp"] / beta_TSS_overlap_df["TSS_size"]
beta_TSS_overlap_df
```

```python
plt.figure(figsize=(6,4))
plt.hist(beta_TSS_overlap_df["overlap_proportion"], bins=30, edgecolor="black")
plt.xlabel("Overlap proportion (Overlap length / TSS length)")
plt.ylabel("Count")
plt.show()
```

```python
# Tag alpha/beta overlap proportion in a single table
alpha_tag = alpha_TSS_overlap_df.assign(usage_group="Enriched in alpha")[["TSS_name","overlap_proportion","usage_group"]]
beta_tag  = beta_TSS_overlap_df.assign(usage_group="Enriched in beta")[["TSS_name","overlap_proportion","usage_group"]]
ovl = pd.concat([alpha_tag, beta_tag], axis=0)

# Keep best overlap per TSS (in case multiple peaks overlap)
ovl_best = ovl.sort_values("overlap_proportion", ascending=False).drop_duplicates("TSS_name")

# Merge and add overlap category
diff = pd.concat([alpha_tss, beta_tss], ignore_index=True)
merged = diff.merge(ovl_best[["TSS_name","overlap_proportion"]], on="TSS_name", how="left")
merged["overlap_cat"] = np.where(merged["overlap_proportion"].fillna(0) > 0.5, ">50% overlap", "â‰¤50% overlap")
```

```python
merged["annotation"] = merged["TSS_name"].str.endswith("_A").map({True: "Annotated", False: "Unannotated"})

order_groups = ["Upregulated in Alpha","Upregulated in Beta"]

# counts for >50% by annotation
hi = (merged.query("overlap_cat == '>50% overlap'")
            .groupby(["significance","annotation"])["TSS_name"]
            .nunique()
            .unstack(fill_value=0)
            .reindex(order_groups, fill_value=0))

# counts for â‰¤50% 
lo = (merged.query("overlap_cat == 'â‰¤50% overlap'")
            .groupby("significance")["TSS_name"]
            .nunique()
            .reindex(order_groups, fill_value=0))

# combine into one table 
plot_df = pd.DataFrame({
    ">50% overlap\n(Annotated TSS)":   hi.get("Annotated", pd.Series(0, index=order_groups)),
    ">50% overlap\n(Unnnotated TSS)": hi.get("Unannotated", pd.Series(0, index=order_groups)),
    "â‰¤50% overlap":       lo
}, index=order_groups)


colors = ["#aec7e8", "#1f77b4", "#d9d9d9"]

ax = plot_df.plot(kind="bar", stacked=True, figsize=(4,2.5), color=colors, width=0.7)


for container in ax.containers:
    ax.bar_label(container, label_type='center', fontsize=8, color="black")

ax.set_ylabel("Number of TSSs", fontsize=10)
ax.set_xlabel("")

ax.set_xticklabels([
    "Alpha-\nenriched",
    "Beta-\nenriched"], rotation=0, ha="center", size = 10)
ax.set_yticklabels(ax.get_yticklabels(), size=8)

ax.legend(
    frameon=False,
    bbox_to_anchor=(1, 0.5),
    loc='center left',
    title=None,
    fontsize = 10
)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

plt.savefig("figures/ATAC_TSS_overlap.pdf", bbox_inches="tight", dpi=300)
plt.show()
```
