## Get GTF of transcripts with unannotated exons for ORFanage

```python
def split_attributes(df, *attribute_type):
    """split attribute of a gtf into separate columns"""
    try: 
        l_attributes = df.strip(';').split('; ')
        attribute_position = [i for i, x in enumerate(l_attributes) if x.startswith(attribute_type)]
        attribute = l_attributes[attribute_position[0]].split(' ')[1].strip('"')
        return attribute
    except Exception:
        return np.nan
```

```python
# Load gtf of all tx
gtf_column_names = ['Chromosome','source','feature','Start','End','score','Strand','frame','attribute']
gtf_df = pd.read_csv( '../data_processed/curated_HIT.gtf',sep='\t', header=None, names=gtf_column_names)
gtf_df.loc[:, 'transcript_id'] = gtf_df['attribute'].apply(split_attributes, args=("transcript_id",))
gtf_df.head()
```

```python
# Keep only the tx containing non-overlapping exons
unmapped_tx_gtf = gtf_df[gtf_df['transcript_id'].isin(nonoverlapping_exons["transcript_id"])].copy()
```

```python
unmapped_tx_gtf["transcript_id"].nunique()
```

```python
# Add transcript features (needed for ORFanage)

# Group by transcripts to get boundaries
transcripts = (
    unmapped_tx_gtf.groupby("transcript_id")
    .agg({
        "Chromosome": "first",
        "source": "first",
        "Start": "min",
        "End": "max",
        "Strand": "first",
        "attribute": "first"
    })
    .reset_index()
)

# Add the columns for transcript entries
transcripts["feature"] = "transcript"
transcripts["score"] = "."
transcripts["frame"] = "."
transcripts = transcripts[[
    "Chromosome", "source", "feature", "Start", "End", "score", "Strand", "frame", "attribute", "transcript_id"
]]

# add transcript rows to original df
unmapped_tx_gtf = pd.concat([transcripts, unmapped_tx_gtf], ignore_index=True)

# Sort 
unmapped_tx_gtf = unmapped_tx_gtf.sort_values(["Chromosome", "Start"]).reset_index(drop=True)
```

```python
unmapped_tx_gtf = unmapped_tx_gtf.drop(columns=["transcript_id"])
unmapped_tx_gtf.head()
```

```python
# Save gtf
unmapped_tx_gtf.to_csv('../data_processed/ORFanage/unannotated_exons_for_orfanage.gtf', sep='\t', header=False, index=False, quoting=csv.QUOTE_NONE)
```

```python
# Mark non-overlapping exons 
orfanage_output_with_stop["is_non_overlapping"] = orfanage_output_with_stop.apply(
    lambda row: (
        (row["Chromosome"], row["Start"], row["End"], row["transcript_id"], row["Strand"]) in nonoverlapping_exons
    ) if row["feature"] == "exon" else False,
    axis=1
)

orfanage_output_with_stop["is_unannotated"] = orfanage_output_with_stop.apply(
    lambda row: (
        (row["Chromosome"], row["Start"], row["End"], row["transcript_id"], row["Strand"]) in unannotated_exons
    ) if row["feature"] == "exon" else False,
    axis=1
)

orfanage_output_with_stop
```

```python
# Mark predicted CDS that match exon with at least one end
exons = orfanage_output_with_stop[(orfanage_output_with_stop["feature"] == "exon") & (orfanage_output_with_stop["is_non_overlapping"] == True)].copy()
exon_start_keys = set(zip(exons["Chromosome"],
                          exons["Start"],
                          exons["transcript_id"],
                          exons["Strand"]))
exon_end_keys   = set(zip(exons["Chromosome"],
                          exons["End"].astype(int),
                          exons["transcript_id"],
                          exons["Strand"]))

# Flag CDS rows whose start OR end matches an exon boundary
orfanage_output_with_stop["CDS_matches_exon"] = orfanage_output_with_stop.apply(
    lambda row: (
        (row["Chromosome"], row["Start"], row["transcript_id"], row["Strand"]) in exon_start_keys
        or
        (row["Chromosome"], row["End"],   row["transcript_id"], row["Strand"]) in exon_end_keys
    ) if row["feature"] == "CDS" else False,
    axis=1
)
```

```python
cds_df = orfanage_output_with_stop[(orfanage_output_with_stop["feature"] == "CDS") & (orfanage_output_with_stop["CDS_matches_exon"] == True)]

orfanage_output_with_stop["is_stop_in_cds"] = orfanage_output_with_stop.apply(
    lambda row: any(
        (row["Start"] >= c["Start"]) and (row["End"] <= c["End"])
        for _, c in cds_df[cds_df["transcript_id"] == row["transcript_id"]].iterrows()
    ) if row["feature"] == "stop_codon" else False,
    axis=1
)
```

```python
orfanage_output_with_stop.query("feature == 'stop_codon'").groupby("transcript_id")["is_stop_in_cds"].all().sum()
```

```python
# Dont have stop codon in CDS
no_stop_codon = orfanage_output_with_stop.groupby("transcript_id")["is_stop_in_cds"].any().loc[lambda x: ~x].index
no_stop_codon
```

```python
df = orfanage_output_with_stop.copy()

# Ensure integer coords (avoids weird comparisons)
df["Start"] = df["Start"].astype(int)
df["End"]   = df["End"].astype(int)

# 1) stop codons that are inside CDS
stops = df.query("feature == 'stop_codon' and is_stop_in_cds")[
    ["transcript_id","Strand","Start","End"]
]

# 2) CDS blocks
cds = df.query("feature == 'CDS'")[
    ["transcript_id","Strand","Start","End"]
]

# 3) Merge stops with CDS of same transcript/strand; suffixes keep names unambiguous
m = stops.merge(cds, on=["transcript_id","Strand"], suffixes=("_stop","_cds"))

# 4) Keep only the CDS exon that actually contains the stop codon
m = m[(m["Start_cds"] <= m["Start_stop"]) & (m["End_stop"] <= m["End_cds"])].copy()

# 5) Compute length within that CDS up to the stop (strand-aware)
m["cds_len_until_stop"] = np.where(
    m["Strand"] == "+",
    m["Start_stop"] - m["Start_cds"],
    m["End_cds"] - m["End_stop"]
)

m["cds_block_len"] = m["End_cds"] - m["Start_cds"] + 1

# fraction of that CDS block until the stop
m["cds_fraction_until_stop"] = m["cds_len_until_stop"] / m["cds_block_len"]

# Find whether that CDS exon is terminal 
# + strand terminal CDS = max End; - strand terminal CDS = min Start
idx_plus  = cds.loc[cds["Strand"] == "+"].groupby("transcript_id")["End"].idxmax()
idx_minus = cds.loc[cds["Strand"] == "-"].groupby("transcript_id")["Start"].idxmin()
terminal  = pd.concat([cds.loc[idx_plus], cds.loc[idx_minus]])[["transcript_id","Start","End"]]
terminal_keys = set(map(tuple, terminal.values))  # (transcript_id, Start, End)

m["is_terminal_cds"] = list(zip(m["transcript_id"], m["Start_cds"], m["End_cds"]))
m["is_terminal_cds"] = m["is_terminal_cds"].isin(terminal_keys)

# Final columns
result = m[[
    "transcript_id","Strand","Start_cds","End_cds","Start_stop","End_stop","cds_len_until_stop","cds_fraction_until_stop","is_terminal_cds"
]].sort_values("cds_len_until_stop")

result
```

```python
classes = ["Annotated exon","Non-overlapping exon"]
groups = [dat.loc[dat["class"]==c,"mean"].values for c in classes]

fig, ax = plt.subplots(figsize=(3,3))
parts = ax.violinplot(groups, showmedians=False, showextrema=False)
 
colors = ["grey", '#4AB0A7'] 
for pc, color in zip(parts['bodies'], colors):
    pc.set_facecolor(color)
    pc.set_edgecolor(color)
    pc.set_alpha(0.5)
for i, (vals, color) in enumerate(zip(groups, colors), start=1):
    lo, hi = np.min(vals), np.max(vals)
    med    = np.median(vals)
    ax.vlines(i, lo, hi, color=color, lw=1)                 # vertical range
    ax.hlines([lo, hi], i-0.08, i+0.08, color=color, lw=1)  # caps
    ax.hlines(med,   i-0.1, i+0.1, color=color, lw=2)     # median

ax.set_xticks([1,2])
ax.set_xticklabels(["Annotated\nexons", "Novel\nexons"])
ax.tick_params(axis='x', labelsize=12)  
ax.tick_params(axis='y', labelsize=10) 
ax.set_ylabel("Mean GERP score", fontsize=12)
plt.tight_layout()
fig.savefig("figures/gerp_violinplot.pdf", bbox_inches='tight', dpi=300)
plt.show()
```

```python
# compare each novel exon to the annotated average score for that gene
ann = (
    annotated_gerp.loc[
        (annotated_gerp["covered"] > 0) & annotated_gerp["mean"].notna(),
        ["Gene name", "mean"]
    ]
    .groupby("Gene name", as_index=False)["mean"]
    .mean()
    .rename(columns={"mean": "annotated_mean"})
)

# get all non-overlapping exons
nov = non_overlapping_gerp.loc[
    (non_overlapping_gerp["covered"] > 0) & non_overlapping_gerp["mean"].notna(),
    ["Gene name", "mean"]
].copy()
nov.rename(columns={"mean": "nonoverlap_mean"}, inplace=True)

# merge annotated means with corresponding non-overlapping exons
merged = nov.merge(ann, on="Gene name", how="left")

high_conservation = merged[merged["nonoverlap_mean"] >= merged["annotated_mean"]].copy()

print(f"{len(high_conservation)} non-overlapping exons have equal or higher conservation than their annotated counterparts.")
```
