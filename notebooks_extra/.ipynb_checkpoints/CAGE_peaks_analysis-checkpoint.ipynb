# Investigating correlation between CAGE quantification and transcript-level RNA-seq quantifications

```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
pd.set_option("display.max_columns", None)
```

```python
tpm_df = pd.read_csv('../data_processed/TSS_info_w_lift.tsv', sep='\t')
tpm_df.head()
```

```python
combined_df = pd.read_csv( '../data_processed/curated_HIT.tsv', sep='\t')
combined_df = combined_df.drop(columns=["exons", "matched_to", "unannotated_exons", 
                                         "summarized_category", "Gene type"])
combined_df.head()
```

```python
# Add CAGE quantification
combined_df = combined_df.merge(
    tpm_df[['transcript_id', 'high_tpm']],
    how="left",
    on="transcript_id"
).rename(columns={'high_tpm': 'CAGE_tpm'})
combined_df.head()
```

```python
combined_df["TSS_name"].nunique()
```

```python
# add the sum of expression of all the transcripts originating from the same TSS
f = lambda x: x.sum()
combined_df['sum_TSS_TPM_ref'] = combined_df.groupby('peak_name')['mean_TPM_ref'].transform(f)
```

```python
# Log transform expression values
correlation = combined_df[(combined_df['CAGE_tpm'] != 0) & (~combined_df["mean_TPM_ref"].isna())].copy()
f = lambda x: np.log2(x)
correlation['CAGE log2 TPM'] = correlation['CAGE_tpm'].apply(f)
correlation['RNA-seq log2 TPM'] = correlation['sum_TSS_TPM_ref'].apply(f)
correlation.head()
```

```python
# correlation between expression of the TSS by CAGE and the sum of the transcript sharing the TSS
stats.pearsonr(correlation['CAGE log2 TPM'],correlation['RNA-seq log2 TPM'])
```

```python
from matplotlib import font_manager
import matplotlib as mpl

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

font_path = "../data_raw/fonts/LatinmodernmathRegular.otf"  
font_manager.fontManager.addfont(font_path)

mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
font = {'family' : 'Latin Modern Math',
       'weight' : 'normal',
       'size'   : 10}

mpl.rc('font', **font)
mpl.rc('lines', lw=1, c='k')
sns.set_context("notebook", font_scale=1, rc={"axes.linewidth": 1,
                                              "lines.linewidth": 1, 
                                              "font.size": 10,
                                              "axes.titlesize": 10,
                                              "axes.labelsize": 12,
                                              "legend.fontsize": 10,
                                              "xtick.labelsize": 10,
                                              "ytick.labelsize": 9})
```

```python
sns.jointplot(x="CAGE log2 TPM", y="RNA-seq log2 TPM", data=correlation,
              kind="hist", bins=80, xlim=(-2,10), ylim=(-2,10), cmap="Blues", height=3)
```

```python
g = sns.jointplot(x="CAGE log2 TPM", y="RNA-seq log2 TPM", data=correlation, kind="kde", xlim=(-2,10), ylim=(-2,10), fill=True, cmap="Blues", height=3)
g.ax_joint.xaxis.set_major_locator(ticker.MultipleLocator(5))
g.ax_joint.yaxis.set_major_locator(ticker.MultipleLocator(5))
g.set_axis_labels(r'CAGE log$_2$(TPM)', r'RNA-seq log$_2$(TPM)')
```

```python
g.fig.savefig("../notebooks/figures/correlation_plot.pdf", format="pdf", bbox_inches="tight")
```
