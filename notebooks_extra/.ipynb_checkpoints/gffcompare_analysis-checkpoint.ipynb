## Repeating comparison of hg19

Ran command: gffcompare -r gencode.v34lift37.annotation.gtf -o hg19 Human_Islet_Transcriptome_v2.2.sorted.gtf

By default, 100bp range allowed from terminal exons of reference transcripts.

```python
gtf_column_names = ['chromosome','source','feature','start','end','score','strand','frame','attribute']
hg19_gffcompare_annotation = pd.read_csv( 'hg19.annotated.gtf',
                           sep='\t', header=None, names=gtf_column_names)
hg19_gffcompare_annotation.head()
```

```python
# filter the transcripts (remove exons)
hg19_gffcompare_annotation = hg19_gffcompare_annotation[hg19_gffcompare_annotation['feature'] == 'transcript']
```

```python
def split_attributes(df, *attribute_type):
    """split attribute of a gtf into separate columns"""
    try: 
        l_attributes = df.strip(';').split('; ')
        attribute_position = [i for i, x in enumerate(l_attributes) if x.startswith(attribute_type)]
        attribute = l_attributes[attribute_position[0]].split(' ')[1].strip('"')
        return attribute
    except Exception:
        return np.nan
```

```python
# split the attributes
hg19_gffcompare_annotation.loc[:, 'transcript_id'] = hg19_gffcompare_annotation['attribute'].apply(split_attributes, args=("transcript_id",))
hg19_gffcompare_annotation.loc[:, 'gene_id'] = hg19_gffcompare_annotation['attribute'].apply(split_attributes, args=("gene_id",))
hg19_gffcompare_annotation.loc[:, 'class_code'] = hg19_gffcompare_annotation['attribute'].apply(split_attributes, args=("class_code",))
```

```python
hg19_gffcompare_annotation.head()
```

### What is the distribution of class codes?

```python
hg19_gffcompare_annotation["class_code"].value_counts()
```

```python
hg19_gffcompare_annotation["class_code"].value_counts().plot(kind='bar')
plt.xlabel("Class Code")
plt.ylabel("Count")
plt.title("GFFCompare Class Code Distribution")
plt.xticks(rotation=0)
plt.show()
```

```python
def get_transcript_type(df):
    """return the similarities between the compared transcripts based on the gffcompare code"""
    class_code = df['class_code']
    gene_id = df['gene_id']
    if class_code == '=':
        transcript_type = "GENCODE annotated transcripts"
    elif class_code in ['k','o']:
        transcript_type = "Unannotated transcripts with annotated and unannotated exons"
    elif class_code in['c','j','e']:
        transcript_type = "Unannotated transcripts with all exons annotated (splice variants)"
    elif class_code in ['x','u','i','y','p','s']:
        if gene_id.startswith('ENSG'):
            transcript_type = "Unannotated transcripts with annotated and unannotated exons"
        else:
            transcript_type = "Unannotated transcripts without any annotated exon (unannotated genes)"
    elif class_code in ['m', 'n']:
        transcript_type = "Unannotated transcripts with retained introns"
    else:
        transcript_type = 'None'
    return transcript_type
```

```python
hg19_gffcompare_annotation['comparison_result'] = hg19_gffcompare_annotation.apply(get_transcript_type, axis=1)
```

```python
hg19_gffcompare_annotation.head()
```

```python
crosstab = pd.crosstab(
    hg19_gffcompare_annotation["class_code"],
    hg19_gffcompare_annotation["comparison_result"]
)
crosstab = crosstab.loc[crosstab.sum(axis=1).sort_values(ascending=False).index]

color_list = ['#00897B', '#3949AB', '#FFA000', '#BA68C8', '#E53935'] 
colors = {key: color_list[i % len(color_list)] for i, key in enumerate(crosstab.columns)}

crosstab.plot(
    kind='bar',
    stacked=True,
    color=[colors[col] for col in crosstab.columns]
)

plt.xlabel("Class Code")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="hg19 comparison results")
plt.tight_layout()
plt.show()
```

```python
# Which transcripts are duplicated in gffcompare annotation?
dup_ids = hg38_gffcompare_annotation['transcript_id'][hg38_gffcompare_annotation['transcript_id'].duplicated()]
duplicated_transcripts = hg38_gffcompare_annotation[hg38_gffcompare_annotation['transcript_id'].isin(dup_ids)].sort_values(by='transcript_id').drop(columns=["attribute"])

# Save to TSV
duplicated_transcripts.to_csv('duplicated_transcripts.tsv', sep='\t', index=False)
duplicated_transcripts
```

```python
# Show only instances of transcripts with + and -
strand_conflicts = duplicated_transcripts.groupby("transcript_id").filter(
    lambda g: set(g["strand"]) == {'+', '-'} and len(g) == 2)
strand_conflicts
```

```python
# Show only instances of transcripts with different chromosomes
chrom_conflicts = duplicated_transcripts.groupby("transcript_id").filter(
    lambda g: len(g["chromosome"].unique()) > 1 and len(g) == 2)
chrom_conflicts
```

```python
# Show all remainig instances
conflict_ids = set(chrom_conflicts["transcript_id"]) | set(strand_conflicts["transcript_id"])
other_conflicts = duplicated_transcripts[~duplicated_transcripts["transcript_id"].isin(conflict_ids)]
other_conflicts.to_csv('other_conflicts.tsv', sep='\t', index=False)
other_conflicts
```

```python
# Which transcripts have both different strands and different chromosomes
conflict_overlap = set(strand_conflicts["transcript_id"]) & set(chrom_conflicts["transcript_id"])

print("Transcript IDs in both strand and chromosome conflicts:")
for tid in sorted(conflict_overlap):
    print(tid)
```
